```python
import random

#Printing welcome message and instructions
print(r"""
Welcome to Dungeon Crawl 1.0!
Kill all the monsters in the dungeon and claim the epic loot.
If your health points reach 0, you loose.
All monster encounters are randomly generated.
Choose from four different actions every turn: left_attack, center_attack, right_attack, or heal.
Enjoy,
Kosta
---------------------------------""")

print(r"""   ********************************************************************************   """)
print(r"""  *                    /   \              /'\       _                              *  """)
print(r"""  *\_..           /'.,/     \_         .,'   \     / \_                            *  """)
print(r"""  *    \         /            \      _/       \_  /    \     _                     *  """)
print(r"""  *     \__,.   /              \    /           \/.,   _|  _/ \                    *  """)
print(r"""  *          \_/                \  /',.,''\      \_ \_/  \/    \                   *  """)
print(r"""  *                           _  \/   /    ',../',.\    _/      \                  *  """)
print(r"""  *             /           _/m\  \  /    |         \  /.,/'\   _\                 *  """)
print(r"""  *           _/           /MMmm\  \_     |          \/      \_/  \                *  """)
print(r"""  *          /      \     |MMMMmm|   \__   \          \_       \   \_              *  """)
print(r"""  *                  \   /MMMMMMm|      \   \           \       \    \             *  """)
print(r"""  *                   \  |MMMMMMmm\      \___            \_      \_   \            *  """)
print(r"""  *                    \|MMMMMMMMmm|____.'  /\_            \       \   \_          *  """)
print(r"""  *                    /'.,___________...,,'   \            \   \        \         *  """)
print(r"""  *                   /       \          |      \    |__     \   \_       \        *  """)
print(r"""  *                 _/        |           \      \_     \     \    \       \_      *  """)
print(r"""  *                /                               \     \     \_   \        \     *  """)
print(r"""  *                                                 \     \      \   \__      \    *  """)
print(r"""  *                                                  \     \_     \     \      \   *  """)
print(r"""  *                                                   |      \     \     \      \  *  """)
print(r"""  *                                                    \            |            \ *  """)
print(r"""   ********************************************************************************   """)

print(r"""
    You enter the dungeon and look around. It is dark and cold, you sense the presence of great evil. 
    While walking in the shadows, you notice something moving. Quick! Attack!""")

#Bonuses
player_dmg_bonus = 0
player_hp_bonus = 0
player_def_bonus = 0

#starting variables
current_monster_name = "unknown monster"
current_monster_hp = 1
current_player_hp = 20 + player_hp_bonus
current_encounter = -1
current_boss_stage = 0
boss_fight_actions = ["attack", "player_attack = THE_FINAL_BOSS_ATK_DMG"]

#player actions
actions = ["left attack", "center attack", "right attack", "heal"]

#player intro action
intro_action = ["attack"]

#function determining player dmg
def player_dmg_func():
    return random.randint(2, 5) + player_dmg_bonus

#function determining player heal
def player_heal_func():
    return random.randint(2, 5)

#function determining monster dmg
def monster_dmg_func():
    return random.randint(0, 4) - player_def_bonus

#defining pool of monsters
monsters = {1: "worm", 2: "wasp", 3: "mosquito", 4: "rat", 5: "bat", 6: "spider", 7: "snake", 8: "soldier", 9: "ooze", \
10: "troll", 11: "wolf", 12: "zombie", 13: "alligator", 14: "ghoul", 15: "imp", 16: "vampire", 17: "warrior", 18: "deamon", \
19: "berserker", 20: "witch", 21: "giant", 22: "cyclop", 23: "centaur", 24: "griffin", 25: "yeti", 26: "werewolf", 27: "dragon"}

#pool of monster health values
monster_health = {"worm": "3", "wasp": "3", "mosquito": "3", "rat": "4", "bat": "4", "spider": "4", "snake": "5", "soldier": "5", "ooze": "5", \
"troll": "7", "wolf": "7", "zombie": "7", "alligator": "8", "ghoul": "8", "imp": "8", "vampire": "10", "warrior": "10", "deamon": "10", \
"berserker": "11", "witch": "11", "giant": "11", "cyclop": "13", "centaur": "13", "griffin": "13", "yeti": "14", "werewolf": "14", "dragon": "15"}

#pool of monster adjectives
monster_adj = {1: "inhuman", 2: "voiceless", 3: "sadistic", 4: "hideous", 5: "nameless", 6: "voiceless", 7: "dreadful", 8: "formidable", \
9: "fearsome", 10: "huge", 11: "deadly", 12: "classic", 13: "horrible", 14: "bloody", 15: "evil", 16: "soulless", 17: "hideous", 18: "hungry", \
19: "insane", 20: "deformed", 21: "bloodthirsty", 22: "unnatural", 23: "mighty", 24: "corrupted", 25: "outerdimensional", 26: "gigantic", \
27: "berserk", 28: "heartless", 29: "murderous", 30: "violent", 31: "enormous", 32: "strange-looking", 33: "powerful", 34: "shadowy", 35: "mysterious", \
36: "massive", 37: "tireless", 38: "supernatural", 39: "threatening", 40: "hostile"}

#1. generate list of monster encounters
fight_encounters = []
for i in range(0, 10):
    encounter = random.randint(1,27)
    fight_encounters.append(encounter)

#2. create a list with the 10 monster types
monster_types = []
for i in fight_encounters:
    monster_types.append(monsters[i])

#3. create a list with the 10 monster health values
monster_hp = []
for i in monster_types:
    monster_hp.append(monster_health[i])
monster_hp = [int(i) for i in monster_hp]

#4.1 random list of 10 adj & converting to monster types
monster_adj_values = []
for i in range(1, 11):
    adj_value = random.randint(1,40)
    monster_adj_values.append(adj_value)
monster_names = []
for i in monster_adj_values:
    monster_names.append(monster_adj[i])

#4.2 creating monster full name (adj + type)
monster_full_names = zip(monster_names, monster_types)
monster_full_names = list(monster_full_names) 
monster_full_names2 = []
for sublist in monster_full_names:
    for item in sublist:
        monster_full_names2.append(item)

#5 monster_info - dict with names and int(hp) of all monsters
monster_full_names3 = []
for i in range(0, len(monster_full_names2), 2):
    monster_full_names3.append(monster_full_names2[i] + " " + monster_full_names2[i+1])
monster_info = dict(zip(monster_full_names3, monster_hp))
# Values = HP | Keys = Name

#function tracking encounters
def get_new_encounter(enc_value):
    global current_encounter
    global current_monster_name
    global current_monster_hp
    if enc_value < 10:
        cur_monst_name = list(monster_info.keys())[enc_value]
        cur_monst_hp = list(monster_info.values())[enc_value]
        enc_value += 1
        current_monster_name = cur_monst_name
        current_monster_hp = cur_monst_hp
        current_encounter = enc_value
    if enc_value == 1:
        print("""\nAll the monsters coming your way are demonic looking and possessed. After you killed
        the Unknown Monster, you notice that its soul escaped its corrupted body and gave you a
        thumbs up. You feel unease, but continue fighting.""")
    if enc_value == 2:
        print("""\nBefore killing the last monster, it starts laughing and says 'THE_FINAL_BOSS is going
         to revenge my death.' You torture the monster and get this information about THE_FINAL_BOSS:
         THE_FINAL_BOSS is the overlord of this place.
         THE_FINAL_BOSS has a special power!
         THE_FINAL_BOSS is aware of your presence""")
    if enc_value == 3:
        global player_hp_bonus
        print("\nYou notice a small room titled 'Laundry' you enter and find THE_FINAL_BOSS's epic armor.")
        print(""" ===!=====!====!===!===!=====!===!===!====""")
        print(r"""  /`\   /`\  |~| |~|  /)=I=(\  /`'''`\   """)
        print(r""" |   `'`   | | | | |  |  :  | |   :   |  """)
        print(r""" '-|     |-' )/\ )/\  |  T  \ '-| : |-'  """)
        print(r"""   |     |  / \// \/  (  |\  |  '---'    """)
        print(r"""   '.___.'  \_/ \_/   |  |/  /           """)
        print(r"""                      |  /  /            """)               
        print(r"""                      |  \ /             """)
        print(r"""                      '--'`              """)
        player_hp_bonus = 10 
        print("Your max health is now 30!")
    if enc_value == 4:
        print("\nYou hear distant screams of monsters being tortured and an evil laugh.")
    if enc_value == 5:
        print("""\nYou find a note that reveals some of THE_FINAL_BOSS's secret plan.
        THE_FINAL_BOSS has the power to ulter reality.
        THE_FINAL_BOSS is a demonic dragon.
        THE_FINAL_BOSS he can code.""")
    if enc_value == 6:
        global player_dmg_bonus
        print("\nThe last monster forgot its sword:")
        print("""                /\                                          """)
        print("""    /vvvvvvvvvvvv \--------------------------------------,  """)
        print("""    `^^^^^^^^^^^^ /====================================='   """)
        print("""                \/                                          """)
        player_dmg_bonus = 2 
        print("Your attacks are doing 2 extra damage!")
    if enc_value == 7:
        print("\nThe dungeon is getting darker and darker. It is hard to breathe.")
    if enc_value == 8:
        print("""You overhear two monsters talking and get some information about
        THE_FINAL_BOSS's secret plan:
        THE_FINAL_BOSS has a few lines of code he will use against you!
        THE_FINAL_BOSS wants to change THE_FINAL_BOSS_ATK_DMG to 1,000,000!
        THE_FINAL_BOSS_FIGHT is very close.
        """)
    if enc_value == 9:
        global player_def_bonus
        print("\nYou find a shield and take it:")
        print("""    |`-._/\_.-`|    """)
        print("""    |    ||    |    """)
        print("""    |___o()o___|    """)
        print("""    |__((<>))__|    """)
        print("""    \   o\/o   /    """)
        print("""     \   ||   /     """)
        print("""      \  ||  /      """)
        print("""       '.||.'       """)
        print("""         ``         """)
        player_def_bonus = 1
        print("You now take less damage!")
    if enc_value == 10:
        start_boss_fight()

#function serves as a tutorial & intro
def player_intro_action(intro_action):
    choice = input("Attack by writing 'attack' in the input field: ").lower()
    while choice not in intro_action:
        print("\nAction not recognised.\n")
        choice = input("Attack by writing 'attack' in the input field: ").lower()
    if choice == "attack":
        print(r"""
        You attacked . . . the mirror of the front door of the dungeon. Damaged by your attack, 
        the whole door falls back and breaks into a thousand pieces. Some pieces hit THE_FINAL_BOSS's car,
        which explodes, creating an even bigger and louder shockwave in the dungeon. All of the monsters 
        in the dungeoun are now pissed off and coming to kill you!""")
        print("Before realizing that you may have screw up big time:")
        player_action(actions)

#function asking for input
def player_action(actions):
    print(current_monster_name.title(), "charges to attack you!")
    choice = input("What are you going to do? (actions: left attack, center attack, right attack, or heal) You choose to: ").lower()
    while choice not in actions:
        print("\nPlease, choose one of the above mentioned actions.\n")
        choice = input("What are you going to do? (actions: left attack, center attack, right attack, or heal) You choose: ").lower()
    global current_player_hp
    global current_monster_hp
    if choice == "left attack":
        player_attacks_monster(current_monster_hp)
    if choice == "center attack":
        player_attacks_monster(current_monster_hp)
    if choice == "right attack":
        player_attacks_monster(current_monster_hp)
    if choice == "heal":
        player_heals(current_player_hp)

#function for player healing
def player_heals(value1):
    global current_player_hp
    player_heal_func()
    player_heal = player_heal_func()
    value1 += player_heal
    if value1 + player_hp_bonus > 20 + player_hp_bonus:
        value1 = 20 + player_hp_bonus
        print("\nYou have reached your maximum health points and cannot heal more!")
    current_player_hp = value1
    print("\nYou healed, and now have", value1, "health points")
    monster_attacks_player(current_player_hp)
    
#function for player attacks on monster
def player_attacks_monster(value2):
    global current_monster_hp
    player_dmg_func()
    player_dmg = player_dmg_func()
    value2 -= player_dmg
    current_monster_hp = value2
    print("\nYou attacked", current_monster_name.title(), "with", player_dmg, "attack, it has", value2, "health left")
    if value2 <= 0:
        print("You slayed", current_monster_name.title())
        get_new_encounter(current_encounter)
        player_action(actions)
    else:
        monster_attacks_player(current_player_hp)
    
#function for monster attacks on player
def monster_attacks_player(value3):
    global current_player_hp
    monster_dmg_func()
    monster_dmg = monster_dmg_func()
    value3 -= monster_dmg
    current_player_hp = value3
    print(current_monster_name.title(), "attacked you back with", monster_dmg,".", "You have", value3, "health left")
    if current_player_hp <= 0:
        print("Game over!", current_monster_name.title(), "killed you!")
    else:
        player_action(actions)

def start_boss_fight():
    global current_monster_name
    global current_monster_hp
    print("""You approach the THE_FINAL_BOSS's office door, suddenly you started hearing BOSS_MUSIC.
        While you were fighting, the THE_FINAL_BOSS was writing code, he changed couple of things:
        THE_FINAL_BOSS_HP = 1,000,000
        THE_FINAL_BOSS_DEF = 1,000,000
        player_heal_action was disabled
        
        There is fire everywhere, you will take 2 damage every turn.
        THE_FINAL_BOSS makes a dramatic cinematic appearance:
        """)
    print(r"""   .:'                                  `:.   """)
    print(r"""  ::'                                    `::  """)
    print(r""" :: :.                                  .: :: """)
    print(r"""  `:. `:.             .             .:'  .:'  """)
    print(r"""   `::. `::           !           ::' .::'    """)
    print(r"""       `::.`::.    .' ! `.    .::'.::'        """)
    print(r"""         `:.  `::::'':!:``::::'   ::'         """)
    print(r"""         :'*:::.  .:' ! `:.  .:::*`:          """)
    print(r"""        :: HHH::.   ` ! '   .::HHH ::         """)
    print(r"""       ::: `H TH::.  `!'  .::HT H' :::        """)
    print(r"""       ::..  `THHH:`:   :':HHHT'  ..::        """)
    print(r"""       `::      `T: `. .' :T'      ::'        """)
    print(r"""         `:. .   :         :   . .:'          """)
    print(r"""           `::'               `::'            """)
    print(r"""             :'  .`.  .  .'.  `:              """)
    print(r"""             :' ::.       .:: `:              """)
    print(r"""             :' `:::     :::' `:              """)
    print(r"""              `.  ``     ''  .'               """)
    print(r"""               :`...........':                """)
    print(r"""               ` :`.     .': '                """)
    print(r"""                `:  `''''  :'                 """)
    print(r"""                                              """)
    current_monster_name = "THE_FINAL_BOSS"
    current_monster_hp = 1000000
    player_boss_fight_action(boss_fight_actions)

#function tracking boss stage
def get_new_boss_stage(boss_stage):
    global current_boss_stage
    global current_player_hp
    if boss_stage < 10:
        boss_stage += 1
        current_boss_stage = boss_stage
    if boss_stage == 1:
        current_player_hp -= 2
        print("The fire damages you, you now have", current_player_hp, "health points left.")
        print("""\n
        THE_FINAL_BOSS is coding, but had an uncommon Git Bash Error, which dalayed his progress
        Update: THE_FINAL_BOSS is currently 10% ready with his code.""")
        player_boss_fight_action(boss_fight_actions)
    if boss_stage == 2:
        current_player_hp -= 2
        print("The fire damages you, you now have", current_player_hp, "health points left.")
        print("""\n
        THE_FINAL_BOSS is still coding, however he made a typo
        Update: THE_FINAL_BOSS is currently 19% ready with his code.""")
        player_boss_fight_action(boss_fight_actions)
    if boss_stage == 3:
        current_player_hp -= 2
        print("The fire damages you, you now have", current_player_hp, "health points left.")
        print("""\n
        THE_FINAL_BOSS is still coding, but forgot to activate conda and had to copy
        and paste his code, which lost him some time
        Update: THE_FINAL_BOSS is currently 34% ready with his code.""")
        player_boss_fight_action(boss_fight_actions)
    if boss_stage == 4:
        current_player_hp -= 2
        print("The fire damages you, you now have", current_player_hp, "health points left.")
        print("""\n
        THE_FINAL_BOSS wasted a bunch of time because of a badly named index he wrote.
        Update: THE_FINAL_BOSS is currently 45% ready with his code.
        """)
        player_boss_fight_action(boss_fight_actions)
    if boss_stage == 5:
        current_player_hp -= 2
        print("The fire damages you, you now have", current_player_hp, "health points left.")
        print("""\n
        THE_FINAL_BOSS drinked 4 coffees and did super much progress!
        Update: THE_FINAL_BOSS is currently 78% ready with his code.
        """)
        player_boss_fight_action(boss_fight_actions)
    if boss_stage == 6:
        current_player_hp -= 2
        print("The fire damages you, you now have", current_player_hp, "health points left.")
        print("""\n
        THE_FINAL_BOSS realized that he used commands from Python 2.7 and this fight is in
        Python 3.8, his progress was delayed.
        Update: THE_FINAL_BOSS is currently 76% ready with his code.
        """)
        player_boss_fight_action(boss_fight_actions)
    if boss_stage == 7:
        current_player_hp -= 2
        print("The fire damages you, you now have", current_player_hp, "health points left.")
        print("""\n
        THE_FINAL_BOSS lost some time trying to modify sets, but also did some progress.
        Update: THE_FINAL_BOSS is currently 84% ready with his code.
        """)
        player_boss_fight_action(boss_fight_actions)
    if boss_stage == 8:
        current_player_hp -= 2
        print("The fire damages you, you now have", current_player_hp, "health points left.")
        print("""\n
        THE_FINAL_BOSS accidently forgot to create a Git commit and pulled from Git, which lost
        him some progress. However, he quickly rewrote his code.
        Update: THE_FINAL_BOSS is currently 88% ready with his code.
        """)
        player_boss_fight_action(boss_fight_actions)
    if boss_stage == 9:
        current_player_hp -= 2
        print("The fire damages you, you now have", current_player_hp, "health points left.")
        print("""\n
        THE_FINAL_BOSS has successfully finished coding! THE_FINAL_BOSS_ATK_DMG is now "1,000,000"!
        THE_FINAL_BOSS laughs at you! He spends a couple of minutes flexing and talking trash to you.
        Finally, at the end of his turn he attacks you! 
        ...
        ...
        ...
        An error appears, the type(THE_FINAL_BOSS_ATK_DMG) was a string. THE_FINAL_BOSS panics and
        quickly changes it to an integer, but this gives you enough time for one last attack!
        """)
        player_boss_fight_action(boss_fight_actions)
    if current_player_hp <= 0:
        print("""\n
        The fire kills you!
        THE_FINAL_BOSS laughs at you!
        Game over!
        """)

#function asking for input in boss fight
def player_boss_fight_action(boss_fight_actions):
    if current_boss_stage < 9:
        choice = input("What are you going to do? (actions: attack) You choose to: ").lower()
        while choice not in boss_fight_actions:
            print("\nPlease, choose one of the above mentioned actions.\n")
            choice = input("What are you going to do? (actions: attack) You choose: ").lower()
        if choice == "attack":
            print("\nYou attacked THE_FINAL_BOSS, but it did not do anything")
            get_new_boss_stage(current_boss_stage)
    if current_boss_stage == 9:
        choice = input("What are you going to do? (actions: attack OR player_attack = THE_FINAL_BOSS_ATK_DMG) You choose to: ")
        while choice not in boss_fight_actions:
            print("\nPlease, choose one of the above mentioned actions.\n")
            choice = input("What are you going to do? (actions: attack OR player_attack = THE_FINAL_BOSS_ATK_DMG) You choose: ")
        if choice == "attack":
            print("\nYou attacked THE_FINAL_BOSS, but it did not do anything")
            current_player_hp = 0
            print("""\n
            You failled to take advantage of the situation and were killed by the THE_FINAL_BOSS.
            Game over!
            """)
        if choice == "player_attack = THE_FINAL_BOSS_ATK_DMG":
            print("""\n
            You change your attack to 1,000,000 and hit THE_FINAL_BOSS.
            The attack is so powerful that it killes THE_FINAL_BOSS instantaneously.
            You won!
            Congratulations, you killed all the monsters in the dungeon.
                 _________________     
                //   EPIC  LOOT   \    
               //__________________\   
               ||-------| |--------|   
               ||       |ยก|        |   
               ||       ---        |   
               ||                  |   
               ||__________________|   
            
            You open the chest with epic loot and find this:
            A gif of a doggo sliding on ice: gph.is/2FH30LZ
            Thanks for playing!
            Kosta
            
            The art in this game was not created by me!
            Check asciiart.website and .eu websites and support the creators!""")

if current_monster_name == "unknown monster":
    player_intro_action(intro_action)
```
